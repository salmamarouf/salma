```python
yaml_content = """
train: dataset/train/images
val: dataset/train/images  # مؤقتًا نستخدم نفس صور التدريب كـ validation

nc: 4  # عدد الكلاسات
names: ['stop', 'turnleft', 'turnright', 'parking']
"""

with open('dataset.yaml', 'w') as f:
    f.write(yaml_content)

print("✅ تم إنشاء ملف dataset.yaml بنجاح")
```

```python
import os

# نتأكد إن مجلد الليبلات موجود
os.makedirs('/content/dataset/train/labels', exist_ok=True)

# ملفات labels - كل واحدة بتشير لكلاس مختلف
labels = {
    'stop.txt': '0 0.5 0.5 0.4 0.4',
    'parking.txt': '1 0.5 0.5 0.4 0.4',
    'turnleft.txt': '2 0.5 0.5 0.4 0.4',
    'turnright.txt': '3 0.5 0.5 0.4 0.4',
}

# نكتب الملفات
for filename, content in labels.items():
    with open(f'/content/dataset/train/labels/{filename}', 'w') as f:
        f.write(content)

print("✅ تم إنشاء ملفات الـ labels بنجاح.")
```

```python
!ls /content/dataset/train/labels
```

```python
from ultralytics import YOLO

# تحميل موديل YOLOv8n (خفيف وسريع)
model = YOLO('yolov8n.pt')

# تدريب الموديل على البيانات المخصصة (4 صور فقط حاليًا)
model.train(
    data='dataset.yaml',  # ملف إعداد البيانات
    epochs=30,            # عدد الـ epochs
    imgsz=640             # حجم الصور
)
```

```python
model.train(data="dataset.yaml", epochs=30, imgsz=640)
```
